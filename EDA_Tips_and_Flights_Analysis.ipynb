```python
#  Distplot
```

```python
sns.distplot(df3[df3['Survived'] == 0]['Age'])
```

```python
sns.displot(df3[df3['Survived'] == 0]['Age'], kind='kde')
sns.displot(df3[df3['Survived'] == 1]['Age'], kind='kde')
```

```python
import matplotlib.pyplot as plt
sns.kdeplot(df3[df3['Survived'] == 0]['Age'], label='Did not survive')
sns.kdeplot(df3[df3['Survived'] == 1]['Age'], label='Survived')

plt.title('Age Distribution by Survival Status (KDE)')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
sns.distplot(df3[df3['Survived'] == 0]['Age'], hist = False)
sns.distplot(df3[df3['Survived'] == 1]['Age'], hist = False)
```

```python
# Heatmap
```

```python
pd.crosstab(df3['Pclass'], df3['Survived'])
```

```python
sns.heatmap(pd.crosstab(df3['Pclass'], df3['Survived']))
```

```python
pd.crosstab(df['Day'], df['Gender'])
```

```python
sns.heatmap(pd.crosstab(df['Day'], df['Gender']))
```

```python
df3.groupby('Pclass')['Survived'].mean()*100
```

```python
(df3.groupby('Pclass')['Survived'].mean()*100).plot(kind='bar')
```

```python
df3.groupby('Embarked')['Survived'].mean()*100
```

```python
non_survivors = df3[df3['Survived'] == 0]
percentage_non_survivors = non_survivors.groupby('Pclass').size() / df3.groupby('Pclass').size() * 100

print(percentage_non_survivors)
```
